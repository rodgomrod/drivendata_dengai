Antes de comenzar vamos a echar un vistazo a la variable respuesta. Como vemos tiene una distribución normal para valores cercanos a cero pero una larga cola. Esto es normal, habra pocas semanas donde se ha resgistrado muchas alarmas y pocas donde ha habido un alto número de alarmas.

```python
plt.figure(figsize=(15,7)) 
sns.distplot(train_labels)
```

Una estrategia que suele dar buenos resultado es normalizar esta cola utilizando, por ejemplo, el logaritmo de uno más equis (1). Se observa como va pareciendose cada vez más a una distribución normal y esto facilita la tarea de regresión a los algoritmos matemáticos.

# Temperaturas al S.I.

Antes de continuar, vamos a transformar las temperaturas al Sistema Internacional. Esto nos va a permitir tener las mediciones en la misma escala para facilitar las imputaciones y las posibles decisiones del futuro algoritmo estadístico.

```python
temp_c_cols = ['station_avg_temp_c', 'station_max_temp_c', 'station_min_temp_c']
train_features[temp_c_cols] += 273.15
test_features[temp_c_cols] += 273.15
```

Ahora tenemos las temperaturas en la misma escala. Podemos estudiar imputar los valores faltantes de una temperatura con la otra, aunque según la documentación en **DrivenData** las temperaturas `reanalysis` son simplemente "temperaturas" y las `station` son las temperaturas del aire.

```python
train_features[['reanalysis_avg_temp_k', 'station_avg_temp_c']].head(10)
```

Si realizamos la diferencia entre `reanalysis_avg_temp_k` y `station_avg_temp_c` observamos el valor máximo, vemos que es solo de ~ 3.47 º K, por lo que una imputación por estos valores puede ser una buena primera aproximación.

```python
pd.Series(train_features['reanalysis_avg_temp_k'] - train_features['station_avg_temp_c']).max()
```

# BenchMark

Se entrenará un primer modelo de XGBoost con todos los datos sin separar por ciudad para establecer un *BenchMark*

```python
from sklearn.model_selection import StratifiedKFold, RandomizedSearchCV
from sklearn.externals import joblib
import xgboost as xgb
```

```python
train_features.columns
```

```python
X_train = train_features.loc[:, [c for c in train_features.columns if c not in  ['total_cases',
                                                                                'week_start_date']]]
y_train = train_features.loc[:,'total_cases']
print(X_train.shape, y_train.shape)
```

Hacemos una búsqueda aleatoria de los mejores parametros para este caso

```python
params = {'max_depth': np.arange(2, 10, 1),
          'metric': ['mae'],
          'n_estimators': [300],
          'eta': np.arange(0.01, 0.1, 0.01),
          'colsample_bytree':np.arange(0.1, 1.1, 0.1),
          'n_jobs': [1],
          'seed': [42],
          'gamma': np.arange(0, 0.9, 0.1),
          'max_bin': np.arange(2, 256, 2),
          'gamma':np.arange(0, 10, 0.1),
          'alpha':np.arange(0, 1.1, 0.1),
          'lambda':np.arange(0,1.1, 0.1),
          'max_delta_step':np.arange(0, 11, 1)
         }

xgb_model = xgb.XGBRegressor()

rscv = RandomizedSearchCV(xgb_model,
                          params,
                          verbose=1,
                          random_state=42,
                          scoring='neg_mean_absolute_error',
                          n_iter=60,
                          cv = 3,
                          n_jobs=-1)
```

```python
%%time
rscv.fit(X_train, y_train)
```

```python
print(rscv.best_params_,'\n')
print(rscv.best_score_)
```

Seleccionamos los mejores parámetros para entrenar el modelo XGBoost final

```python
params = rscv.best_params_
xgb_model = xgb.XGBRegressor(**params)

xgb_model.n_estimators = 10000
xgb_model.n_jobs = 1
```

```python
X_test = test_features.loc[:, [c for c in test_features.columns if c not in  ['total_cases',
                                                                              'week_start_date']]]
```

Hacemos cross-validación con 5 Folds

```python
k = 5
train_ids = X_train.index
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)
```

```python
counter1 = 1
y_preds = np.zeros(X_test.shape[0])
be = 0
stack = pd.DataFrame()
stack2 = pd.DataFrame()
for train_index, test_index in skf.split(train_ids, y_train):
    print('Fold {}\n'.format(counter1))

    X_fit, X_val = X_train.iloc[train_index, :], X_train.iloc[test_index, :]
    y_fit, y_val = y_train.iloc[train_index], y_train.iloc[test_index]

    xgb_model.fit(X_fit,
                  y_fit,
                  eval_set=[(X_val, y_val)],
                  verbose=100,
                  early_stopping_rounds=50)

    y_preds += xgb_model.predict(X_test)

    be += xgb_model.best_score
    
    del X_fit
    del X_val
    del y_fit
    del y_val
    del train_index
    del test_index
    gc.collect()

    counter1 += 1
    
y_preds = y_preds / k
```

```python
sub = pd.read_csv('../data/submission_format.csv', index_col=[0,1,2])
```
